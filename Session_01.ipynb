# Basic Python semantics
## Variables
**Variables are pointers**. In other languages, variables can be thought of as *buckets* of data. However, in Python, variables are simply pointers to objects. So when you write:


```python
x = 2
```

you are essentially defining a *pointer* named **x** that points to some an object containing the value 4. Note one consequence of this: because Python variables just point to various objects, there is no need to "declare" the variable, or even require the variable to always point to information of the same type!

This is the sense in which people say Python is ***dynamically-typed***: variable names can point to objects of any type (numeric arrays, strings, lists…).

```python
x = 1         # x is an integer
x = 'hello'   # now x is a string
x = [1, 2, 3] # now x is a list
```


**Important consequence**: If we have two variable names pointing to the same *mutable* object, then modifying one will change the other as well! For instance, take a look at the following code

```python
x = [1, 2, 3, 4]
y = x
print('x is ', x)
print('y is ', y)

x.append(5) # change value  of x

print('x now is ', x)
print('y now is ', y)
```

This behavior might seem confusing if you're wrongly thinking of variables as buckets that contain data. But if you're correctly thinking of variables as pointers to objects, then this behavior makes sense.

Note also that if we use "=" to assign another value to x, this will not affect the value of y – assignment is simply a change of what object the variable points to:


```python
x = 'something else'
print(y)  # y is unchanged
```
